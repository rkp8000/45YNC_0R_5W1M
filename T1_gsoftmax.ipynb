Tests for softmax function that accepts infinite gain.

```python
import numpy as np
from ntwk import gsoftmax
```

```python
z = np.array([
    [0, 1, -1],
    [-5, -5, -5],
    [1, 1, 1],
    [0, 10, 0],
    [100, 0, 100.],
])

for g in [1, 2, -2, 10000, np.inf, -10000, -np.inf]:
    print(f'g = {g}:')
    for z_row, soft_row in zip(z, gsoftmax(z, g)):
        print(f'{z_row} -> {soft_row}')
    print()
```
